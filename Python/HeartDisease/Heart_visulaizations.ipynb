```python
# Scatter plot
plt.scatter(dh['age'], dh['chol'])
  
# Adding Title to the Plot
plt.title("Scatter Plot")
  
# Setting the X and Y labels
plt.xlabel('age')
plt.ylabel('chol')
  
plt.show()
```

```python
# Bar chart 
plt.bar(dh['age'], dh['chol'])
  
plt.title("Bar Chart")
  
# Setting the X and Y labels
plt.xlabel('age')
plt.ylabel('chol')
  
# Adding the legends
plt.show()

```

```python
sns.scatterplot(x='chol', y='target', data=dh,)
plt.show()
```

```python
sns.lineplot(x='age', y='target', data=dh)
plt.show()
```

```python
sns.histplot(x='target', data=dh, hue='target')
plt.show()
```
