# 1. CREATE SUBSETS

```python
#Since mean of the column area is 12.84 which is close to 13 subsets created w.r.t mean for column "area"
```

```python
notAffected = df1[df1['area']==0]
```

```python
notAffected
```

```python
mostlyAffected = df1[df1['area'] > 13 ]
```

```python
mostlyAffected
```

```python
partiallyAffected = df1[(df1['area']  <= 13 ) & (df1['area']  > 0 )  ]
```

```python
partiallyAffected
```

# 2. Merged two subsets

```python
merge = [partiallyAffected,mostlyAffected]

mergedDataset = pd.concat(merge)
```

```python
mergedDataset.shape
```
